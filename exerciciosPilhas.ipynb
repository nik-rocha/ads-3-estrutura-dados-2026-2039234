1) Caixas no Deposito

**Problema**: Um deposito registra caixas identificadas por numeros inteiros. As caixas sao empilhadas
na ordem em que chegam. Sua tarefa e simular a pilha apos algumas operacoes basicas.

**Tarefa**: Empilhe os valores 1, 2, 3 e 4, nessa ordem. Em seguida, remova duas caixas do topo. Ao
final, informe o ultimo valor removido e mostre como a pilha ficou.

**Entrada**: Nao ha entrada via teclado. Os valores devem ser definidos diretamente no codigo.

**Saida**: Seu programa deve exibir o ultimo valor removido e o estado final da pilha

```python
class Stack:
    
    def __init__(self):
        self.deposito = []
        self.removidos = []
        self.ultima_remocao = None
        
    def isEmpty(self):
        return self.deposito == []
    
    def push(self, item):
        self.deposito.append(item)
        
    def pop(self):
        self.ultima_remocao = self.deposito.pop()
        return self.removidos.append(self.ultima_remocao)
    
    def peek(self):
        return self.deposito[len(self.deposito)-1]
    
    def size(self):
        return len(self.deposito)
    
deposito = Stack()
deposito.push(1)
deposito.push(2)
deposito.push(3)
deposito.push(4)

print(f"Pilha atual: {deposito.deposito}")

deposito.pop()
deposito.pop()

print(f"Removidos: {deposito.removidos}")
print(f"Último removido: {deposito.ultima_remocao}")
print(f"Pilha final: {deposito.deposito}")
```

2) Mesa de Livros

**Problema**: Em uma biblioteca, os livros devolvidos sao organizados em uma pilha temporaria antes
de voltarem para as estantes. O bibliotecario quer observar o livro que esta no topo antes e depois
de uma retirada.

**Tarefa**: Empilhe os livros Python, Algoritmos e Estruturas de Dados, nessa ordem. Mostre qual livro
esta no topo, remova um livro e mostre o novo topo.

**Entrada**: Nao ha entrada via teclado. Utilize os titulos indicados no enunciado.

**Saida**: Exiba o topo antes da remocao, o livro removido e o topo apos a remocao.


```python
class Stack:
    
    def __init__(self):
        self.biblioteca = []
        self.ultima_remocao = None
        
    def isEmpty(self):
        return self.biblioteca == []
    
    def push(self, item):
        self.biblioteca.append(item)
        
    def pop(self):
        self.ultima_remocao = self.biblioteca.pop()
        return self.ultima_remocao
    
    def peek(self):
        return self.biblioteca[len(self.biblioteca)-1]
    
    def size(self):
        return len(self.biblioteca)
    
biblioteca = Stack()
biblioteca.push("Python")
biblioteca.push("Algoritmos")
biblioteca.push("Estruturas de Dados")

print(f"Pilha: {biblioteca.biblioteca}")
print(f"Topo inicial:")
print(biblioteca.peek())

biblioteca.pop()

print(f"Removido: {biblioteca.ultima_remocao}")
print(f"Novo topo: {biblioteca.peek()}")
```

3) Palavra Invertida

**Problema**: Em um jogo de linguistica, uma palavra precisa ser mostrada ao contrario. Para isso,
deve-se usar o comportamento LIFO de uma pilha.

**Tarefa**: Dada uma palavra, empilhe seus caracteres e depois desempilhe todos eles para montar a
palavra invertida.

**Entrada**: Uma palavra definida no codigo. Exemplo: python

**Saida**: A palavra invertida.

```python
class Stack:
    
    def __init__(self, palavra:str):
        self.palavra = palavra
        self.arranjo = []
        
    def isEmpty(self):
        return self.arranjo == []
    
    def push(self, item):
        self.arranjo.append(item)
        
    def pop(self):
        return self.arranjo.pop()
    
    def peek(self):
        return self.arranjo[len(self.arranjo)-1]
    
    def size(self):
        return len(self.arranjo)
    
    def reverse(self):
        reversed = ""
        while not self.isEmpty():
            reversed += self.pop()
        return reversed
    
palavra = Stack("python")

for letra in palavra.palavra:
    palavra.arranjo.append(letra)
    
# palavra_invertida = "".join(reversed(palavra.arranjo))
# print(palavra_invertida)

palavra.reverse()
```
